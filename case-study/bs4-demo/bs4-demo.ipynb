```python
import requests
from pprint import pprint
from bs4 import BeautifulSoup
import urllib.request

URL = "https://baike.baidu.com/item/DeepSeek/65368136"
page = requests.get(URL)

pprint(page.text)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
# results = soup.find(id="ResultsContainer")
results = soup.find_all("div", class_="itemWrapper_Wc7NF")
```

### 直觉方式

```python
all_parts = []
for result in results:
    temp_part = []
    for part in str(result).split('>'):
        temp_part.append(part.strip('</dt').strip('</span'))
    
    all_parts.append(temp_part)
```

```python
all_infos = []
for each_part in all_parts:
    temp = []
    for text in each_part:
        if ('=' not in text) and (len(text) > 3):
            temp.append(text)
    all_infos.append(temp)
```
